```python
import pandas as pd
```

```python
occupations_df = pd.read_csv('/content/drive/MyDrive/NLP_Talent_clef_2025/esco_raw_data/occupations_en.csv')
occupations_df.head()
```

```python
occupations_df.info()
```

```python
print(occupations_df.isnull().sum())
```

```python
occupations_df=occupations_df.dropna(subset=['altLabels'])
```

```python
occupations_df.shape
```

```python
similar_title_sets = pd.DataFrame({"jobtitle" : [],
                               "similar_titles" : []})
similar_title_sets.head()
```

```python
similar_title_sets['jobtitle']=occupations_df['preferredLabel']
similar_title_sets['similar_titles']=occupations_df['altLabels']
```

```python
similar_title_sets.head()
```

```python
similar_title_sets['similar_titles'] = similar_title_sets['similar_titles'].apply(lambda x: x.split("\n"))
```

```python
similar_title_sets.head()
```

```python
similar_title_sets['similar_titles'] = similar_title_sets.apply(lambda row: [row['jobtitle']] + row['similar_titles'], axis=1)

similar_title_sets.head()
```

```python
similar_title_sets['similar_titles'] = similar_title_sets['similar_titles'].apply(lambda x: "\n".join(x))
```

```python
similar_title_sets.head()
```

```python
similar_title_sets.to_csv('/content/drive/MyDrive/NLP_Talent_clef_2025/esco_similar_titles/similar_title_sets_english.csv', index=False)
```
